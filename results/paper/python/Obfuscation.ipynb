```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
%matplotlib inline 

matplotlib.rcParams.update({'font.size': 15})

# in seconds
gcc_non_stripped_linux_avg = 44.554
gcc_stripped_linux_avg = 44.555
llvm_linux_avg = 37.724
llvm_obfuscated_linux_avg = 109.64
gcc_non_stripped_hermitux_avg = 44.502
gcc_stripped_hermitux_avg = 44.577
llvm_hermitux_avg = 38.294
llvm_obfuscated_hermitux_avg = 109.924

gcc_glibc_linux_avg = 44.254
gcc_glibc_hermitux_avg = 44.37
gfortran_linux_avg = 43.286
gfortran_hermitux_avg = 43.141

gcc_non_stripped_linux_std = 0.151231977805989
gcc_stripped_linux_std = 0.182528536581726
llvm_linux_std = 0.070269323162941
llvm_obfuscated_linux_std = 0.670906517747113
gcc_non_stripped_hermitux_std = 0.13587576024524
gcc_stripped_hermitux_std = 0.150262732867763
llvm_hermitux_std = 0.302294925895593
llvm_obfuscated_hermitux_std = 1.16465731726833

gcc_glibc_linux_std = 0.061137368096588
gcc_glibc_hermitux_std = 0.078315600829805
gfortran_linux_std = 0.758700643644559
gfortran_hermitux_std = 0.134035650315711

linux_means = [gcc_non_stripped_linux_avg, gcc_stripped_linux_avg, gcc_glibc_linux_avg, llvm_linux_avg, \
               llvm_obfuscated_linux_avg, gfortran_linux_avg]
linux_stdevs = [gcc_non_stripped_linux_std, gcc_stripped_linux_std, gcc_glibc_linux_std, llvm_linux_std, \
                llvm_obfuscated_linux_std, gfortran_linux_std]
hermitux_means = [gcc_non_stripped_hermitux_avg, gcc_stripped_hermitux_avg, gcc_glibc_hermitux_avg, \
                  llvm_hermitux_avg, llvm_obfuscated_hermitux_avg, gfortran_hermitux_avg]
hermitux_stdevs = [gcc_non_stripped_hermitux_std, gcc_stripped_hermitux_std, gcc_glibc_hermitux_std, \
                   llvm_hermitux_std, llvm_obfuscated_hermitux_std, gfortran_hermitux_std]

fig, ax = plt.subplots()
ax.grid()
ax.set_axisbelow(True)

plt.ylabel("NPB IS Class A\nexecution time (s)")
plt.xticks((.5, 3.5, 6.5, 9.5, 12.5, 15.5), ("GCC/\nMusl", "GCC/Musl\nstripped", "GCC/\nGlibC", "Clang", \
                                 "Clang\nobfuscated", "Gfortran"))

l1 = mpatches.Patch(facecolor='skyblue', hatch="//", label='Linux', edgecolor='black')
l2 = mpatches.Patch(facecolor='moccasin', label='HermiTux', hatch="\\\\", edgecolor='black')
plt.legend(handles=[l1, l2])


rects1 = ax.bar([0, 3, 6, 9, 12, 15], linux_means, yerr=linux_stdevs, hatch="//", color=["skyblue"], \
                edgecolor=['black', 'black', 'black', 'black', 'black', 'black'])
rects2 = ax.bar([1, 4, 7, 10, 13, 16], hermitux_means, yerr=hermitux_stdevs, hatch="\\\\", color=["moccasin"], \
                edgecolor=['black', 'black', 'black', 'black', 'black', 'black'])

fig = matplotlib.pyplot.gcf()
fig.subplots_adjust(bottom=0.2)
fig.subplots_adjust(left=0.2)
fig.set_size_inches(7,4)

plt.savefig("Obfuscation.pdf")



```
